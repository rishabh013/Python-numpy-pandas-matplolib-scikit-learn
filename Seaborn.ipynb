# Heatmap

```python
# heatmap is a graphical representation of 2D(two dimensional) data. Each data value represent 
# in a matrix and it has a special colour.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d )
```

```python
globalWarming_df = pd.read_csv("D:\\Assets\\Who_is_responsible_for_global_warming.csv")
globalWarming_df.head()
```

```python
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'],
                                          axis=1).set_index('Country Name')
globalWarming_df
```

```python
plt.figure(figsize=(14,7))
 
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(14,6))
 
sns.heatmap(globalWarming_df, vmin = 0, vmax = 21)
```

```python
plt.figure(figsize=(14,6))
 
sns.heatmap(globalWarming_df, cmap="coolwarm") # cmap = colour change
```

```python
# center parameter 
plt.figure(figsize=(14,6))
 
sns.heatmap(globalWarming_df, cmap="coolwarm", center = 10.0)
```

```python
# annot (annotate) parameter
 
plt.figure(figsize=(14,6))
 
sns.heatmap(globalWarming_df, annot = True)
```

```python
arr_2d
```

```python
annot_arr = np.arange(1,13).reshape(4,3) # create 2D numpy array with 4 rows and 3 columns
annot_arr
```

```python
annot_arr = np.arange(1,13).reshape(4,3) # create 2D numpy array with 4 rows and 3 columns
 
sns.heatmap(arr_2d, annot= annot_arr)
```

```python
# fmp parameter - add text on heatmap cell
 
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(arr_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
```

## Part 10

```python
plt.figure(figsize=(14,6))
annot_kws = {'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
sns.heatmap(globalWarming_df, cmap="coolwarm", annot = True, annot_kws = annot_kws)
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(globalWarming_df, cmap="coolwarm", annot = True, linewidth =4, linecolor ='k')
```
